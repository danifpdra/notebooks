```python
import matplotlib.pyplot as plt
import os
import numpy as np
import statistics 
import math  
```

```python
txt_file_rgb =os.path.join("/home/daniela/workspace/ball_detection/Datasets/train/4cams_sep1/", "rgb.txt")
f_rgb = open(txt_file_rgb, "r")

txt_file_ir = os.path.join("/home/daniela/workspace/ball_detection/Datasets/train/4cams_sep1/", "ir.txt")
f_ir = open(txt_file_ir, "r")

txt_file_mono =os.path.join("/home/daniela/workspace/ball_detection/Datasets/train/4cams_sep1/", "mono.txt")
f_mono = open(txt_file_mono, "r")

txt_file_depth =os.path.join("/home/daniela/workspace/ball_detection/Datasets/train/4cams_sep1/", "depth.txt")
f_depth = open(txt_file_depth, "r")
```

```python
rgb=[]
ir=[]
mono=[]
depth=[]

for line_rgb in f_rgb: 
    n=len(line_rgb)
    timestamp=line_rgb[69:n-5]
#     print(timestamp)
    timestamp=float(timestamp)*10**-3
#     print(timestamp)
    rgb.append(int(timestamp))
    
for line_ir in f_ir: 
    n=len(line_ir)
    timestamp=line_ir[68:n-5]
    print(timestamp)
    timestamp=float(timestamp)*10**-3
    ir.append(int(timestamp))
    
for line_mono in f_mono: 
    n=len(line_mono)
    timestamp=line_mono[70:n-5]
    print(timestamp)
    timestamp=float(timestamp)*10**-3
    mono.append(int(timestamp))
    
for line_depth in f_depth: 
    n=len(line_depth)
    timestamp=line_depth[71:n-5]
    print(timestamp)
    timestamp=float(timestamp)*10**-3
    depth.append(int(timestamp))
```

```python
n=min(len(ir),len(rgb))
# print (n_ir,n_rgb)
# print (rgb)
print(rgb)
```

```python
plt.figure(figsize=(30,20))
plt.scatter(range(0,n), ir[0:n],color= "red",marker= "*", s=10)
plt.scatter(range(0,n),rgb[0:n],color= "blue", marker= "*", s=10)

# naming the x-axis 
plt.xlabel('Frame number') 
# naming the y-axis 
plt.ylabel('Timestamp (ns)') 
# plot title 
plt.title('Timestamp analysis!')

plt.show()
```

```python
desvio=0

for i in range(0,n):
    desvio+=((ir[i]-rgb[i])*10**-3)**2

desvio=math.sqrt(abs(desvio)/n)
print("Error:", "%.2f" % desvio, "ms")
```
